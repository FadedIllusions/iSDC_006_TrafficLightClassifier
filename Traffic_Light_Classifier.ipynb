# 2. Pre-process the Data

After loading in each image, you have to standardize the input and output!

### Input

This means that every input image should be in the same format, of the same size, and so on. We'll be creating features by performing the same analysis on every picture, and for a classification task like this, it's important that **similar images create similar features**! 

### Output

We also need the output to be a label that is easy to read and easy to compare with other labels. It is good practice to convert categorical data like "red" and "green" to numerical data.

A very common classification output is a 1D list that is the length of the number of classes - three in the case of red, yellow, and green lights - with the values 0 or 1 indicating which class a certain image is. For example, since we have three classes (red, yellow, and green), we can make a list with the order: [red value, yellow value, green value]. In general, order does not matter, we choose the order [red value, yellow value, green value] in this case to reflect the position of each light in descending vertical order.

A red light should have the  label: [1, 0, 0]. Yellow should be: [0, 1, 0]. Green should be: [0, 0, 1]. These labels are called **one-hot encoded labels**.

*(Note: one-hot encoding will be especially important when you work with [machine learning algorithms](https://machinelearningmastery.com/how-to-one-hot-encode-sequence-data-in-python/)).*

## Standardize the output

With each loaded image, we also specify the expected output. For this, we use **one-hot encoding**.

* One-hot encode the labels. To do this, create an array of zeros representing each class of traffic light (red, yellow, green), and set the index of the expected class number to 1. 

Since we have three classes (red, yellow, and green), we have imposed an order of: [red value, yellow value, green value]. To one-hot encode, say, a yellow light, we would first initialize an array to [0, 0, 0] and change the middle value (the yellow value) to 1: [0, 1, 0].

## Construct a `STANDARDIZED_LIST` of input images and output labels.

This function takes in a list of image-label pairs and outputs a **standardized** list of resized images and one-hot encoded labels.

This uses the functions you defined above to standardize the input and output, so those functions must be complete for this standardization to work!


````python
def standardize(image_list):
    
    # Empty image data array
    standard_list = []

    # Iterate through all the image-label pairs
    for item in image_list:
        image = item[0]
        label = item[1]

        # Standardize the image
        standardized_im = standardize_input(image)

        # One-hot encode the label
        one_hot_label = one_hot_encode(label)    

        # Append the image, and it's one hot encoded label to the full, processed list of image data 
        standard_list.append((standardized_im, one_hot_label))
        
    return standard_list

# Standardize all training images
STANDARDIZED_LIST = standardize(IMAGE_LIST)
````

````python
# (Optional) Add more image analysis and create more features

def test_color_dom(channel):
    """
        This function is used in conjunction with the color_dominance().
        Essentially, all that this function does is find the largest value
        within the R color channel and the G color channel of and RGB image.
        Those values will then be passed to the color_dominance().
    """
    channel = channel
    # Rows and Cols cropped slightly to reduce noise
    rows = int(channel.size/channel[0].size)-2
    cols = int(channel[0].size)-2
    temp = 0
    
    for row in range(rows):
        for col in range(cols):
            if channel[row][col] > temp:
                temp = channel[row][col]
    return temp

def color_dominance(rgb_image):
    """
        This function compares the largest value from the R channel and
        the largest value from the G channel of an RGB image so as to
        increase the probability of differentiating between a red light
        and a green light.
        
        Originally, student was going to use an HSV image and simply
        compare the hue to differentiate between R, Y, G; but, couldn't
        find any documentation on the bounds of each. Red starts at an
        H of 0... but, the endpoint isn't clear enough for comfort. That,
        and, honestly, student couldn't really figure out how to code such
        a comparison... so, student improvised, adapted, overcame. :)
    """
    masked_image = hsv_crop_and_mask(rgb_image)
    color_image = cv2.cvtColor(masked_image, cv2.COLOR_HSV2RGB)
    
    r_largest = (test_color_dom(color_image[:,:,0]))
    g_largest = (test_color_dom(color_image[:,:,1]))
    
    if r_largest > g_largest:
        return "red"
    else:
        return "green"
    


# Use dominant color along with regional summation to determine label
def region_specifier(rgb_image):
    masked_image = hsv_crop_and_mask(rgb_image)
    h = masked_image[:,:,2]
    
    # Rows and Cols cropped slightly to reduce noise
    rows = int(h.size/h[0].size)-2
    cols = int(h[0].size)-2

    # Determine approximate size of regional sections
    row_third = int(rows/3)
    
    # Define region/section variables
    r_region = row_third
    y_region = row_third*2
    g_region = row_third*3
    
    # Variables to hold sectional sums
    r_sum = 0
    y_sum = 0
    g_sum = 0
    
    # Determine region of light data
    for row in range(rows):
        for col in range(cols):
            # First section, Red Light section
            if row <= r_region:
                r_sum += h[row][col]
            if row > r_region and row <= y_region:
                y_sum += h[row][col]
            if row > y_region and row <= g_region:
                g_sum += h[row][col]

    # Determine light color
    if r_sum > y_sum and r_sum > g_sum:
        return "red"
    if y_sum > r_sum and y_sum > g_sum:
        return "yellow"
    else:
        return "green"
    
    
#print(color_dominance(MISCLASSIFIED[0][0]))
#print(region_specifier(MISCLASSIFIED[0][0]))

# At this point, student realized that, though built, he never implemented the brightness feature in the create_feature().
# And, based on testing of his other two features (color_dominance() and region_specifier()), student doesn't seem to need
# it for classification. However, it /is/ on the rubric. Will student 'fail' because brightness is not specifically used?


# ... student remedied this and implemented brightness, just not in the region specifier function
````

````python
# This function should take in RGB image input
# Analyze that image using your feature creation code and output a one-hot encoded label
def estimate_label(rgb_image):
    
    ## TODO: Extract feature(s) from the RGB image and use those features to
    ## classify the image and output a one-hot encoded label
    
    predicted_label = [0,0,0]
    dominance = color_dominance(rgb_image)
    region = region_specifier(rgb_image)
    
    # Red Light
    if dominance == states[0] and region == states[0]:
        predicted_label = one_hot_encode(states[0])
    
    # Green Light
    if dominance == states[2] and region == states[2]:
        predicted_label = one_hot_encode(states[2])
    
    # Yellow Light
    if region == states[1]:
        predicted_label = one_hot_encode(states[1])
        
    
    
    return predicted_label   
    
#print(estimate_label(mis))
````
